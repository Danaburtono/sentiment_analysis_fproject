```python
# Add new column for sentiment analysis
new_sentiment_df = sentiment_df[['tweet_id','full_text']]
new_sentiment_df.insert(2,"sentiment", '')
new_sentiment_df.head()
```

```python
# Function to split dataframe
def split_dataframe_by_position(df, splits):
    """
    Takes a dataframe and an integer of the number of splits to create.
    Returns a list of dataframes.
    """
    dataframes = []
    index_to_split = len(df) // splits
    start = 0
    end = index_to_split
    for split in range(splits):
        temporary_df = df.iloc[start:end, :]
        dataframes.append(temporary_df)
        start += index_to_split
        end += index_to_split
    return dataframes
```

```python
# Splitting Dataframe into 4 dataFrames
split_dataframes = split_dataframe_by_position(new_sentiment_df, 4)
```

```python
# Save 1st Dataframe to csv file
split_dataframes[0].to_csv('../dana/dana_sentiment_analysis.csv', index=False)
```
